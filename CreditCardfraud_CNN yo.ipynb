```python
import pandas as pd 
import numpy as np
from tensorflow import keras

path ="/content/drive/MyDrive/train.csv"

data = pd.read_csv(path)
```

```python
print(data.shape)
print(data.columns)
print(data.head())
Label = data.train.cvs['Class']
X = data.cvs[:,'V1':'V28']
print(X.shape)
print(Label.shape)
```

```python
print(X.shape)
X1 = X.values # pandas dataframe is converted into numpy 
Datset = np.zeros((X1.shape[0],28,28))

# converting my 1-D datset into 2-D image  to be processed in a  CNN network 
for i in range(X1.shape[0]):
  im = np.zeros((28,28))
  for j in range(28):
    im[j,:]=X1[i,:].reshape(1,-1)
  Datset[i,:,:]=im


    


```

```python
print(Datset.shape) 
# apply the CNN on this and look for the accuracy 
```

```python
CNNModel = keras.models.Sequential()
# input layer 
CNNModel.add(keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
CNNModel.add(keras.layers.MaxPool2D(2,2))

# other layer of convolution 
CNNModel.add(keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
CNNModel.add(keras.layers.MaxPool2D(2,2))

# third layer - > wihtout Maxpooling
CNNModel.add(keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'))
```

```python
CNNModel.summary()

Label = Label.values
```

```python
CNNModel.add(keras.layers.Flatten())
CNNModel.add(keras.layers.Dense(350,activation='relu'))
CNNModel.add(keras.layers.Dense(10))
CNNModel.summary()

CNNModel.compile(optimizer = 'adam',loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])

History = CNNModel.fit(Datset,Label,epochs = 10)

# split the data into 70:30 ratio and chk for train and validation acucracy 
# evaluate 
```
